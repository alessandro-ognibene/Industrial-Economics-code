```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Define the specific probabilities and profits
profitc = 30   # Profit for mid-tier products
profitb = 0     # Profit for low-tier products
profitm = 2.25*profitc #  # Profit for high-tier products

# Range of p_tsmc values from 0 to 1
p_tsmc_values = np.linspace(0, 1, 101)

# Range of profitb values from -10 to profitc (adjust as needed)
profitb_values = np.linspace(-10, profitc, 101)

# Initialize matrices to store payoffs
intel_payoff_invest = np.zeros((len(p_tsmc_values), len(profitb_values)))
intel_payoff_not_invest = np.zeros((len(p_tsmc_values), len(profitb_values)))
intel_payoff_difference = np.zeros((len(p_tsmc_values), len(profitb_values)))

# Calculate payoffs for each combination of p_tsmc and profitb
for i, p_tsmc in enumerate(p_tsmc_values):
    for j, profitb in enumerate(profitb_values):
        # Payoff when Intel invests
        payoff_1_intel_invest = p_tsmc * profitm * (1 - p_tsmc) + (1 - p_tsmc) * (p_tsmc * profitb + (1 - p_tsmc) * profitc)
        
        # Payoff when Intel does not invest
        payoff_1_intel_not_invest = p_tsmc * profitb + (1 - p_tsmc) * profitc
        
        # Store payoffs in respective matrices
        intel_payoff_invest[i, j] = payoff_1_intel_invest
        intel_payoff_not_invest[i, j] = payoff_1_intel_not_invest
        intel_payoff_difference[i, j] = payoff_1_intel_invest - payoff_1_intel_not_invest

# Create a 3D surface plot for the difference in Intel's payoffs
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Create grid of p_tsmc and profitb values
p_tsmc_grid, profitb_grid = np.meshgrid(p_tsmc_values, profitb_values)

# Plot the surface for the difference in payoffs
surf = ax.plot_surface(profitb_grid, p_tsmc_grid, intel_payoff_difference.T, cmap='coolwarm')
ax.set_xlabel('Profits in B')
ax.set_ylabel('p_tsmc (Probability TSMC Invests)')
ax.set_zlabel('Difference in Intel\'s Payoff')
ax.set_title('Difference in Intel\'s Payoff Depending on p_tsmc and profitb')
fig.colorbar(surf, label='Payoff Difference')

# Adjust viewing angle for better perspective
ax.view_init(20, 45)  # Set elevation and azimuth angles
plt.show()
```
